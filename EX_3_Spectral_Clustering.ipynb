```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
A1 = np.zeros((6,6))
A1
```

```python
for i in range(6):
    for j in range(i,6):
        A1[i,j] = input("element %s %s"%(i,j))
```

```python
A1 = A1 + A1.T
A1
```

```python
G1 = nx.Graph(A1)
nx.draw_networkx(G1)
```

```python
D1 = np.zeros((4,4))
for i in range(4):
    for j in range(i+1,4):
        D1[i,j] = input("element %s %s"%(i,j))

D1
```

```python
D1 = D1.T + D1
```

```python
GD1 = nx.Graph(D1)
ND = nx.algorithms.tree.mst.minimum_spanning_tree(GD1)
nx.draw_networkx(ND)
```

```python
nx.draw_networkx(GD1)
```
